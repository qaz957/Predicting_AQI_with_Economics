```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_2010 = pd.read_csv("..\\Cleaned_Data\\2010_Data_clean.csv")
```

```python
data_2010.head()
```

```python
# Remove ',' from String literals in DataFrame
data_2010['Civilian employed population 16 years and over_2010'] = data_2010['Civilian employed population 16 years and over_2010'].str.replace(',','')
data_2010['Agriculture, forestry, fishing and hunting, and mining_2010'] = data_2010['Agriculture, forestry, fishing and hunting, and mining_2010'].str.replace(',','')
data_2010['Construction_2010'] = data_2010['Construction_2010'].str.replace(',','')
data_2010['Manufacturing_2010'] = data_2010['Manufacturing_2010'].str.replace(',','')
data_2010['Wholesale trade_2010'] = data_2010['Wholesale trade_2010'].str.replace(',','')
data_2010['Retail trade_2010'] = data_2010['Retail trade_2010'].str.replace(',','')
data_2010['Transportation and warehousing, and utilities_2010'] = data_2010['Transportation and warehousing, and utilities_2010'].str.replace(',','')
data_2010['Information_2010'] = data_2010['Information_2010'].str.replace(',','')
data_2010['Finance and insurance, and real estate and rental and leasing_2010'] = data_2010['Finance and insurance, and real estate and rental and leasing_2010'].str.replace(',','')
data_2010['Professional, scientific, and management, and administrative and waste management services_2010'] = data_2010['Professional, scientific, and management, and administrative and waste management services_2010'].str.replace(',','')
data_2010['Educational services, and health care and social assistance_2010'] = data_2010['Educational services, and health care and social assistance_2010'].str.replace(',','')
data_2010['Arts, entertainment, and recreation, and accommodation and food services_2010'] = data_2010['Arts, entertainment, and recreation, and accommodation and food services_2010'].str.replace(',','')
data_2010['Other services, except public administration_2010'] = data_2010['Other services, except public administration_2010'].str.replace(',','')
data_2010['Public administration_2010'] = data_2010['Public administration_2010'].str.replace(',','')
```

```python
# Remove industry_2010 column
del data_2010['Industry_2010']
```

```python
data_2010.head()
```

```python
# Convert values in data_2010 to integer type for Regression
data_2010_new = data_2010.copy()
data_2010_new = data_2010_new.astype(int)
```

```python
# Convert values in data_2010 to proportion of total employed
data_2010_new['Agriculture, forestry, fishing and hunting, and mining_2010'] = (data_2010_new['Agriculture, forestry, fishing and hunting, and mining_2010']/data_2010_new['Civilian employed population 16 years and over_2010'])*100
data_2010_new['Construction_2010'] = (data_2010_new['Construction_2010']/data_2010_new['Civilian employed population 16 years and over_2010'])*100
data_2010_new['Manufacturing_2010'] = (data_2010_new['Manufacturing_2010']/data_2010_new['Civilian employed population 16 years and over_2010'])*100
data_2010_new['Wholesale trade_2010'] = (data_2010_new['Wholesale trade_2010']/data_2010_new['Civilian employed population 16 years and over_2010'])*100
data_2010_new['Retail trade_2010'] = (data_2010_new['Retail trade_2010']/data_2010_new['Civilian employed population 16 years and over_2010'])*100
data_2010_new['Transportation and warehousing, and utilities_2010'] = (data_2010_new['Transportation and warehousing, and utilities_2010']/data_2010_new['Civilian employed population 16 years and over_2010'])*100
data_2010_new['Information_2010'] = (data_2010_new['Information_2010']/data_2010_new['Civilian employed population 16 years and over_2010'])*100
data_2010_new['Finance and insurance, and real estate and rental and leasing_2010'] = (data_2010_new['Finance and insurance, and real estate and rental and leasing_2010']/data_2010_new['Civilian employed population 16 years and over_2010'])*100
data_2010_new['Professional, scientific, and management, and administrative and waste management services_2010'] = (data_2010_new['Professional, scientific, and management, and administrative and waste management services_2010']/data_2010_new['Civilian employed population 16 years and over_2010'])*100
data_2010_new['Educational services, and health care and social assistance_2010'] = (data_2010_new['Educational services, and health care and social assistance_2010']/data_2010_new['Civilian employed population 16 years and over_2010'])*100
data_2010_new['Arts, entertainment, and recreation, and accommodation and food services_2010'] = (data_2010_new['Arts, entertainment, and recreation, and accommodation and food services_2010']/data_2010_new['Civilian employed population 16 years and over_2010'])*100
data_2010_new['Other services, except public administration_2010'] = (data_2010_new['Other services, except public administration_2010']/data_2010_new['Civilian employed population 16 years and over_2010'])*100
data_2010_new['Public administration_2010'] = (data_2010_new['Public administration_2010']/data_2010_new['Civilian employed population 16 years and over_2010'])*100
```

```python
x = data_2010_new[['Agriculture, forestry, fishing and hunting, and mining_2010','Construction_2010','Manufacturing_2010','Wholesale trade_2010','Retail trade_2010', 'Transportation and warehousing, and utilities_2010','Information_2010','Finance and insurance, and real estate and rental and leasing_2010', 'Professional, scientific, and management, and administrative and waste management services_2010','Educational services, and health care and social assistance_2010','Arts, entertainment, and recreation, and accommodation and food services_2010','Other services, except public administration_2010','Public administration_2010']]
y = data_2010_new['2010_Median_AQI']
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = pd.DataFrame()
vif_data["feature"] = x.columns
vif_data["VIF"] = [variance_inflation_factor(x.values, i) for i in range(len(x.columns))]
print(vif_data)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(x_train, y_train)
```

```python
print("Intercept: ", mlr.intercept_)
print("Coefficients:")
list(zip(x, mlr.coef_))
```

```python
#Prediction of test set
y_pred_mlr= mlr.predict(x_test)
```

```python
from sklearn import metrics
meanAbErr = metrics.mean_absolute_error(y_test, y_pred_mlr)
meanSqErr = metrics.mean_squared_error(y_test, y_pred_mlr)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(y_test, y_pred_mlr))
print('R squared: {:.2f}'.format(mlr.score(x_test,y_test)*100))
print('Mean Absolute Error:', meanAbErr)
print('Mean Square Error:', meanSqErr)
print('Root Mean Square Error:', rootMeanSqErr)
```
