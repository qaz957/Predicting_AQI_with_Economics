```python
# Set variables for 2010 data 
x = data_10new[['Agriculture, forestry, fishing and hunting, and mining_2010','Construction_2010','Manufacturing_2010','Wholesale trade_2010','Transportation and warehousing, and utilities_2010','Information_2010','Finance and insurance, and real estate and rental and leasing_2010', 'Arts, entertainment, and recreation, and accommodation and food services_2010','Public administration_2010']]
y = data_10new['2010_Binning']

# Use VIF to review collinearity of variables
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = pd.DataFrame()
vif_data["feature"] = x.columns
vif_data["VIF"] = [variance_inflation_factor(x.values, i) for i in range(len(x.columns))]
print(vif_data)
```

```python
# Split dataset into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=100)

# Create and fit DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=4, random_state=2)
clf.fit(x_train, y_train)

#Prediction of test set
y_pred= clf.predict(x_test)

from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# Create visual for Decision Tree
from sklearn.tree import export_text
x1 = ['Agriculture, forestry, fishing and hunting, and mining_2010','Construction_2010','Manufacturing_2010','Wholesale trade_2010','Transportation and warehousing, and utilities_2010','Information_2010','Finance and insurance, and real estate and rental and leasing_2010', 'Arts, entertainment, and recreation, and accommodation and food services_2010','Public administration_2010']
tree_rules = export_text(clf, feature_names=x1)
print(tree_rules)
```

```python
# Print the classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
