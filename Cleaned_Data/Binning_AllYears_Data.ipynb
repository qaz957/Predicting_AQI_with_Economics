```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
AllData = pd.read_csv('AllYears_Data_w_States.csv', delimiter=',',index_col=None) 

AllData.head(10)
```

```python
bins = [10, 30, 50, 70,90]
labels = ['1', '2', '3','4']
AllData['2019_Binning'] = pd.cut(x = AllData['2019_Median_AQI'], bins = bins, labels = labels, include_lowest = True)
AllData['2015_Binning'] = pd.cut(x = AllData['2015_Median_AQI'], bins = bins, labels = labels, include_lowest = True)
AllData['2010_Binning'] = pd.cut(x = AllData['2010_Median_AQI'], bins = bins, labels = labels, include_lowest = True)
```

```python
AllData.head(30)
```

```python
AllData.to_csv('AllYears_Data_Binning.csv', index=False)
```
