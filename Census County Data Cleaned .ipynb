```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
Occupation = pd.read_csv('OCCUPATION BY COUNTY.csv', delimiter=',',index_col=None) 

Occupation.head(10)
```

```python
# DROP NaN VALUES FROM DATA
Drop_NA = Occupation.dropna(subset=['Information'])

Drop_NA.head(10)
```

```python
# DROP ESTIMATE ROWS
Cleaned_Census = Drop_NA[Drop_NA['County_Name'].str.contains('Estimate')==False]

Cleaned_Census.tail()
```

```python
Cleaned_Census.to_csv('Cleaned_Census_Data.csv', index=False)
```
