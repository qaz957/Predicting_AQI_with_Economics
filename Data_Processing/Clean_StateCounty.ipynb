```python
import pandas as pd

AQI_indexes = pd.read_csv("/content/drive/MyDrive/Team 03/Pollution_Data/Median_AQI_byCounty.csv")
AQI_indexes.head(10)
```

```python
# Select out relevant years

AQI_select = AQI_indexes[["State", "County", "2019_Median_AQI", "2015_Median_AQI", "2010_Median_AQI"]]
AQI_select
```

```python
# Change format of county column

for i in range(len(AQI_select)):

  AQI_select["County"].values[i] = AQI_select["County"].values[i] + f" County, {AQI_select['State'].values[i]}"


# Baldwin County, Alabama
```

```python
AQI_select.head(10)
```

```python
# Drop State column and change county column name

QI_cleaned = AQI_select.drop("State", axis=1)


AQI_cleaned = AQI_cleaned.rename(columns={"County": "County_Name"})

AQI_cleaned.head()
```

```python
# Save to CSV
AQI_cleaned.to_csv("/content/drive/MyDrive/Team 03/Pollution_Data/Median_AQI10-15-19.csv")
```
