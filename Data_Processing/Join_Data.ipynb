```python
import pandas as pd
import numpy as np

Median_AQI = pd.read_csv("/content/drive/MyDrive/Team 03/Pollution_Data/Median_AQI10-15-19.csv")

Econ_2010 = pd.read_csv("/content/drive/MyDrive/Team 03/Economic_Data/Census_Data/Cleaned_Occupation_2010_Data.csv")

Econ_2015 = pd.read_csv("/content/drive/MyDrive/Team 03/Economic_Data/Census_Data/Cleaned_Occupation_2015_Data.csv")

Econ_2019 = pd.read_csv("/content/drive/MyDrive/Team 03/Economic_Data/Census_Data/Cleaned_Occupation_2019_Data.csv")
```

```python
Econ_2010.columns
```

```python
# Join first 2 dfs
joined_data1 = pd.merge(Econ_2010, Econ_2015, how="inner", on = ["County_Name"])
joined_data1.columns
```

```python
# Join next 2 dfs
joined_data2 = pd.merge(joined_data1, Econ_2019, how="inner", on = ["County_Name"])
joined_data2
```

```python
# Join final 2 dfs
joined_data3 = pd.merge(joined_data2, Median_AQI, how="inner", on = ["County_Name"])
```

```python
# Change 'N' to null values
joined_data4 = joined_data3.replace('N', np.nan)
joined_data4.columns
```

```python
joined_data4.isnull().sum()
```

```python
#Remove empty columns
joined_data5 = joined_data4.drop(["PERCENT ALLOCATED", "PERCENT ALLOCATED_x", "PERCENT ALLOCATED_y", 'Unnamed: 0'], axis=1)
joined_data5.isnull().sum()
```
